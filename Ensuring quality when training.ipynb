## Facebook training dataset

```python
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn import datasets, linear_model
from sklearn import svm, metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error, r2_score, roc_curve, precision_score, recall_score, accuracy_score, roc_auc_score
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
facebook = pd.read_csv('facebook_test_in_class.csv')
facebook.head()
#Change categorical variables to numeric 
cleanup_nums = {"type": {"Photo": 1, "Link": 2, "Status": 3, "Video": 4}}
facebook.replace(cleanup_nums, inplace=True)
facebook.head()
```

```python
#Choose train and target variables
facebook_train = facebook [['total_followers', 'type', 'paid']]
facebook_target = facebook[['comment', 'like', 'share']]
```

```python
regr1 = linear_model.LinearRegression()
regr2 = linear_model.LinearRegression()
regr3 = linear_model.LinearRegression()

comment_fit = regr1.fit(facebook_train, facebook_target['comment'])
like_fit = regr2.fit (facebook_train, facebook_target['like'])
share_fit = regr3.fit (facebook_train, facebook_target['share'])

y_pred_comment = cross_val_predict(comment_fit, facebook_train, facebook_target['comment'], cv=3)
y_pred_like = cross_val_predict(like_fit, facebook_train, facebook_target['like'], cv=3)
y_pred_share = cross_val_predict(share_fit, facebook_train, facebook_target['share'], cv=3)

MSE_comment = mean_squared_error (facebook_target['comment'], y_pred_comment)
MAE_comment = mean_absolute_error (facebook_target['comment'], y_pred_comment)
Median_AE_comment = median_absolute_error (facebook_target['comment'], y_pred_comment)
R2_comment = r2_score (facebook_target['comment'], y_pred_comment)

MSE_like = mean_squared_error (facebook_target['like'], y_pred_like)
MAE_like = mean_absolute_error (facebook_target['like'], y_pred_like)
Median_AE_like = median_absolute_error (facebook_target['like'], y_pred_like)
R2_like = r2_score (facebook_target['like'], y_pred_like)

MSE_share = mean_squared_error (facebook_target['share'], y_pred_share)
MAE_share = mean_absolute_error (facebook_target['share'], y_pred_share)
Median_AE_share = median_absolute_error (facebook_target['share'], y_pred_share)
R2_share = r2_score (facebook_target['share'], y_pred_share)

print("MSE of comment:", MSE_comment)
print ("MAE_comment:", MAE_comment)
print ("Median_AE_comment:", Median_AE_comment)
print ("R2_comment", R2_comment)

print("MSE_like:", MSE_like)
print ("MAE_like:", MAE_like)
print ("Median_AE_like:", Median_AE_like)
print ("R2_like", R2_like)

print("MSE_share:", MSE_share)
print ("MAE_share:", MAE_share)
print ("Median_AE_share:", Median_AE_share)
print ("R2_share", R2_share)
```

```python
#Split the model into test and training set
X_train, X_test, y_train, y_test = train_test_split(facebook_train, facebook_target, test_size=0.3, random_state=0)
X_train.shape, y_train.shape
X_test.shape, y_test.shape  
X_train.head()
```

```python
#Fit the model
knn = KNeighborsClassifier()
knn.fit(X_train, y_train) 

#Predict new model
y_pred = knn.predict(X_test)
mean_squared_error(y_test, y_pred)
```
